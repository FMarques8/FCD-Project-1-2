### Select rows to use and load file to dataframe

```python
# rows with useful information
row_of_interest = [4,6]
row_of_interest += range(8,22)
row_of_interest.pop(-2)

data = pd.read_csv("ine_principais_culturas_agricolas.csv", encoding = 'latin1',
    sep = ';', header = [1,3], skiprows = lambda x: x not in row_of_interest
    )

# print first 5 rows
data.head()
```

```python
# drop row "0" as it doesn't contain useful data
data.drop([0], inplace= True)
data
```

### Replace first column which has region names with respective code number

```python
# adds 2 columns at the end, one with region number and another the region name
# select data from first column and split on the ":" and return into 2 lists
# ['Region']=region_id and ['Region_name']= region name
data[['Region', 'Region_name']] = data[data.columns[0]].str.split(":", 
    expand = True
    )

# create list with region names for later use
region_name=data['Region_name'].to_list()

# replace index with region numbers
data.set_index('Region', inplace = True)

#remove first, and last 2 columns 
data.drop(data.columns[[0, -1, -2]], axis = 1, inplace = True)
data
```

### Fix the headers

```python
# data.columns.get_level_values.to_series() returns Series 
# of first row of the dataframe
row0 = data.columns.get_level_values(0).to_series()

# does the same but for the second row
row1 = list(data.columns.get_level_values(1).to_series())

#row0.mask replaces values if condition is true
# ffill == fillna(method = 'ffill') -> this replaces null values with previous row value
row0 = row0.mask(lambda r: r.str.startswith("Unnamed")).ffill()

# replaces header with correct labels
data.columns = [row0, row1]
data
```

### Now load and clean the IPMA data

```python
import os #to handle several files

path = os.getcwd()+"//IPMA//" #saves csv directory to variable

station_region = []
data_2 = pd.DataFrame()

counter = 0 # For dict of dict of dataframes
dict_data = {}
# iterates over files in IPMA/
with os.scandir(path) as itr:
    for entry in itr:
        if not entry.name.endswith(".xlsx"):
            continue
        tmp_meta = pd.read_excel(path+entry.name, 'meta', header=None) # reads 'meta' xlsx sheet which has region name
        station_region.append(tmp_meta.iloc[2][1]) # adds region to list
        tmp_data = pd.read_excel(path+entry.name, [1,2,3]) # reads remaining sheets, which have data on temperature and precipitation
        dict_data[counter] = tmp_data # adding sheet information to dictionary {counter: data}
        counter +=1

        
sheet_names = ['Min temp', 'Max temp', 'Precipitation']
```
