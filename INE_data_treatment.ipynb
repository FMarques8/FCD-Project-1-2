### Now load and clean the IPMA data

```python
import os #to handle several files

path = os.getcwd()+"//IPMA//" #saves csv directory to variable

station_region = []
data_2 = pd.DataFrame()

counter = 0 # For dict of dict of dataframes
dict_data = {}
# iterates over files in IPMA/
with os.scandir(path) as itr:
    for entry in itr:
        if not entry.name.endswith(".xlsx"):
            continue
        tmp_meta = pd.read_excel(path+entry.name, 'meta', header=None) # reads 'meta' xlsx sheet which has region name
        station_region.append(tmp_meta.iloc[2][1]) # adds region to list
        tmp_data = pd.read_excel(path+entry.name, [1,2,3]) # reads remaining sheets, which have data on temperature and precipitation
        dict_data[counter] = tmp_data # adding sheet information to dictionary {counter: data}
        counter +=1

        
sheet_names = ['Min temp', 'Max temp', 'Precipitation']
```

### Reduce region name size and group regions by relative position geographically

```python
station_region_fix = [s.replace(" / ", " ").replace(", ", " ") for s in station_region] #replaces separators with spaces
station_region_fix=[s.split(' ')[0] for s in station_region_fix[1:]] #splits regions on spaces
station_region_fix.insert(0, station_region[0]) #adds first word of each region to new list
station_region_fix[5] = 'São Miguel' # manually replace 'São' with 'São Miguel'
station_region_fix[3] = 'Castelo Branco'
print(station_region_fix)
```

```python
import random

# pt: portugal 1:continente 11:norte 16:centro 17:lisboa 18:alentejo 15:algarve 2: Açores 3: Madeira

# region that each station belongs to
station_dict = {'Sub-região do Alto Tâmega': ['PT', '1', '11'],
            'Santarém': ['PT', '1', '18'], 'Setúbal': ['PT', '1', '16'],
            'Castelo Branco': ['PT', '1', '16'], 'Terceira': ['PT', '2'],    
            'São Miguel': ['PT', '2'], 'Madeira': ['PT', '3'],
            'Lisboa': ['PT', '1', '17'], 'Porto': ['PT', '1', '11'],
            'Coimbra': ['PT', '1', '16'], 'Faro': ['PT', '1', '15'],
            'Évora': ['PT', '1', '18'], 'Beja': ['PT', '1', '18'],
            'Guarda': ['PT', '1', '11'], 'Portalegre': ['PT', '1', '16'],
            'Bragança': ['PT', '1', '11']
            }

# each station in region, manually so it has the same order as previous dataframes
region_dict = {'PT':[], '1':[], '11':[], '16':[], '17':[], '18':[], '15':[], '2':[], '3':[]}

# iterates over keys in region_dict then iterates over keys in station_dict
# if station not in region dict (avoids duplicates) and region in that station then adds to region list
# this dictionary is to help with randomly chosen station
for i in list(region_dict.keys()):
    for j in list(station_dict.keys()):
        if i in station_dict[j] and j not in region_dict[i]:
            region_dict[i].append(j)
```

### Convert dict of DataFrames to separate DataFrames 

```python
# year_level = pd.DataFrame.from_dict(dict_data[1][1])['year'].to_list() # year as list for renaming level 0 later
# months_level = pd.DataFrame.from_dict(dict_data[1][1]).columns.get_level_values(0)[1:].to_list() # months for renaming level 1
# data_type_level = ['tmin', 'tmax', 'prec'] # data type to rename level 2

station_data = {list(station_dict.keys())[x]: dict_data[x] for x in range(len(station_dict.keys()))}

data2_by_station = {}
for i in range(len(list(dict_data.keys()))):
    data2_by_station[station_region_fix[i]] = {'tmin': dict_data[i][1], 'tmax': dict_data[i][2], 'prec': dict_data[i][3]}

# replacing Nan values with 0
for i in list(data2_by_station.keys()):
    for j in list(data2_by_station[i].keys()):
        data2_by_station[i][j] = data2_by_station[i][j].fillna(0)
```

```python
# information about tmin of 'Castelo Branco' station
data2_by_station['Castelo Branco']['tmin']
```

### Grouping the data by region with the average value or choosing a random station to represent the region

```python
#initialize dictionary of dictionaries of the different dataframes
data2_average = {j:{'tmin': [], 'tmax': [], 'prec': []} for i in list(station_dict.keys()) for j in station_dict[i]}

# Grouping the region data using the average values
# list form to then be easier to group
for i in list(station_dict.keys()):
    for j in station_dict[i]:
        data2_average[j]['tmin'].append(data2_by_station[i]['tmin']) 
        data2_average[j]['tmax'].append(data2_by_station[i]['tmax'])  
        data2_average[j]['prec'].append(data2_by_station[i]['prec']) 


year_col = data2_average['PT']['tmin'][0]['year'] # column with year number, for later

# test by printing 'PT' values
data2_average['PT']
```

### Now calculating the average values

```python
for i in list(data2_average.keys()):
    tmp_df_tmin = data2_average[i]['tmin'][0] #initializes temporary dataframe with first dataframe from list
    tmp_df_tmax = data2_average[i]['tmax'][0]
    tmp_df_prec = data2_average[i]['prec'][0]

    for j in range(1,len(data2_average[i]['tmin'])): #can be any information since they have the same length
        tmp_df_tmin += data2_average[i]['tmin'][j] #adds every dataframe from region to the first dataframe
        tmp_df_tmax += data2_average[i]['tmax'][j]
        tmp_df_prec += data2_average[i]['prec'][j]

    data2_average[i]['tmin']=tmp_df_tmin/len(data2_average[i]['tmin']) # replaces list of dataframes in region with a single dataframe with the average values
    data2_average[i]['tmax']=tmp_df_tmin/len(data2_average[i]['tmax'])
    data2_average[i]['prec']=tmp_df_tmin/len(data2_average[i]['prec'])
    
    
    # as all values were summed up, this results in the year also being added together, simply replacing with the original year values fixes this
    data2_average[i]['tmin']['year'] = year_col
    data2_average[i]['tmax']['year'] = year_col
    data2_average[i]['prec']['year'] = year_col
    
    data2_average[i]['tmin'] = data2_average[i]['tmin'].set_index('year').T
    data2_average[i]['tmax'] = data2_average[i]['tmax'].set_index('year').T
    data2_average[i]['prec'] = data2_average[i]['prec'].set_index('year').T
```

### Printing resulting tmin dataframes for average values analysis

```python
new_idx = list(data2_average.keys())

# creating list of tuples with new levels for index
# this is the multi index we're going to use
idx_arr = [(i,j) for i in new_idx for j in list(data2_average['PT']['tmin'].index)] 
idx_arr = pd.MultiIndex.from_tuples(idx_arr)
idx_arr
```

```python
tmin_df = pd.DataFrame()
tmax_df = pd.DataFrame()
prec_df = pd.DataFrame()

for i in list(data2_average.keys()):
    tmin_df = pd.concat([tmin_df, data2_average[i]['tmin']], axis=0) # concatenates all dataframes down to one for each type of info
    tmax_df = pd.concat([tmax_df, data2_average[i]['tmax']], axis=0) 
    prec_df = pd.concat([prec_df, data2_average[i]['prec']], axis=0)
```

```python
tmin_df = pd.concat([tmin_df], keys = new_idx, names = ['Region']) # add new index level 
tmax_df = pd.concat([tmax_df], keys = new_idx, names = ['Region'])
prec_df = pd.concat([prec_df], keys = new_idx, names = ['Region'])

tmin_df.index = idx_arr # fixes index level names
tmax_df.index = idx_arr
prec_df.index = idx_arr
```

```python
final_df2 = pd.concat([tmin_df, tmax_df, prec_df], axis=1, keys= ['tmin', 'tmax', 'prec'])
final_df2 = final_df2.unstack(level=1).reorder_levels([1,0,2], axis=1) 

final_df2
```
